```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Ã–rnek: sin, cos

```python
x = np.linspace(0,5,50)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, axes = plt.subplots(2,1)

axes[0].plot(x, y1)
axes[0].set_title("SinÃ¼s fonksiyonu")

axes[1].plot(x, y2, color = "green")
axes[1].set_title("KosinÃ¼s fonksiyonu")

plt.tight_layout()
```

## 2. Ã–rnek: for dÃ¶ngÃ¼sÃ¼

```python
x = np.linspace(1,11,100)
```

```python
y_values = [np.sin(x),np.cos(x), np.tan(x), (x*10-20)]
```

```python
fig, axis = plt.subplots(2,2,figsize=(10,8))

for i, ax, in enumerate(axis.flat):

    ax.plot(x,y_values[i])
    ax.set_title(f"{i+1}. grafik")
    ax.grid(True)

plt.tight_layout()
```
