```python
import numpy as np
import matplotlib.pyplot as plt
```

## Veri olu≈üturma

```python
Example_List = np.arange(40)
np_Ex = np.array(Example_List)
print(np_Ex)
```

```python
Y_List = np.random.randint(10,100,40)
np_Y = np.array(Y_List)
print(np_Y)
```

## Eksenler

```python
my_Figure = plt.figure()

ax_1 = my_Figure.add_axes([.1,.1,1,1]) # .1 = 0.1
ax_2 = my_Figure.add_axes([.8,.8,.3,.3]) # [x, y, x_size, y_size]

ax_1.plot(np_Ex,np_Y, color = "red")
ax_1.set_title("Fig√ºr 1 grafiƒüi")

ax_2.plot(np_Y,np_Ex, color = "blue", marker="o")
```
